```python
import math
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Bank_Churn.csv')  
df2 = pd.read_csv('Bank_Churn_Messy.csv') 
```

```python
df = pd.merge(df1, df2, on='CustomerId')
```

```python
df
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.drop(columns=['Balance_x', 'NumOfProducts_x', 'HasCrCard_x', 'Exited_x', 'Unnamed: 7','Unnamed: 8'], inplace=True)
```

```python
df
```

```python
df
```

```python
df.rename(columns={'Balance_y': 'Balance', 
                   'NumOfProducts_y': 'NumOfProducts',
                  'HasCrCard_y':'HasCrCard',
                  'Tenure_y':'Tenure',
                  'IsActiveMember_y':'IsActiveMember',
                  'Exited_y':'Exited'}, inplace=True)
```

```python
df
```

```python
def conv(value):
    if value == 1:
        return("yes")
    else:
        return("no")
df['Exited'] = df['Exited'].apply(conv)
```

```python
df.drop(columns=['Balance_x', 'NumOfProducts_x', 'HasCrCard_x', 'Exited_x', 'Tenure_x','Unnamed: 7','Unnamed: 8'], inplace=True)
```

```python
df.drop(columns = ['IsActiveMember_x'], inplace = True)
```

```python
gb = df.groupby("Exited").agg({'Exited':"count"})
plt.pie(gb['Exited'], labels = gb.index, autopct = "%1.2f%%", colors=["lightgreen", "tomato"])
plt.title("Percentage churn out")
plt.show()
```

```python
x = sns.countplot( x="Geography", data = df, hue = "Exited", palette = ['orange','limegreen'])

for container in x.containers:
    x.bar_label(container)

plt.title("Customer Churn Distribution Across Countries", color = 'red')
plt.figure(figsize = (3,4))
plt.show()
```

```python
top_ages = df['Age'].value_counts().nlargest(10).sort_values(ascending=True) 

filtered_df = df[df['Age'].isin(top_ages.index)]

plt.figure(figsize=(12, 6))

x = sns.countplot(x="Age", data=filtered_df, hue="Exited", order=top_ages.index, palette = ['royalblue','mediumseagreen'])

for container in x.containers:
    x.bar_label(container)
    
plt.title("Customer Churn Analysis by Top 10 Age Groups", color = 'red')
plt.show()
```
